## `conda` 的安装和使用
### 安装 `miniconda3`
在终端中输入以下命令。
1. `wget https://repo.continuum.io/miniconda/Miniconda3-latest-Linux-x86_64.sh -O ~/miniconda.sh`2. `bash ~/miniconda.sh -b -p $HOME/miniconda3`3. `echo 'export PATH="$PATH:$HOME/miniconda3/bin"' >> ~/.profile.local`4. `source ~/.profile.local`
### 为训练创建一个新的 `conda` 环境
环境名称可以自由选择，为了简化起见，本次训练我们将使用名称 `troll`。
1. `conda create -n troll python=3.6`
使用命令激活 `troll` 环境
2. `source activate troll`
你应该在终端提示符中看到conda环境已激活的提示，例如：`(troll) joro@x2g13:~ $`. 当 `conda` 环境处于活动状态时，之后的所有安装都将仅针对此特定环境。可以使用 `source deactivate` 命令来停用该环境。
### 所需库和模块的安装
使用这个庞大的一行命令安装所需的库和模块：
1. `conda install -c conda-forge gdal satpy jupyter ffmpeg "imageio<2.5" imageio-ffmpeg shapely cartopy scipy rasterio mysql-connector-python matplotlib cartopy scipy pycoast pyorbital`
### 应用软件包更新
PyTroll 包持续开发中。为了完成某些练习，我们需要获取其中一个读取器的更新版本。
1. `pip install git+https://github.com/pytroll/satpy.git`2. `cp /tcenas/scratch/pytroll/seviri_l1b_native.py ~/miniconda3/envs/troll/lib/python3.6/site-packages/satpy/readers/`
