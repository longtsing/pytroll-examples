# 关于区域和pyresample的更多信息
## 摘要
在本教程中，我们将学习如何在pytroll/satpy中定义工作区域，以及如何保存它们以备以后使用。
## 定义区域定义
pytroll/satpy 中的区域通过 pyresample 包进行定义。
定义一个区域，请查看 pyresample 文档：https://pyresample.readthedocs.io/en/latest/geo_def.html

在本教程中，我们将创建一个涵盖瑞典的区域定义，遵循官方的SWEREF99TM标准。
- 应使用的投影为横轴墨卡托投影，参数如下：- 参考经度是15- 缩放因子为 0.9996
瑞典被包含在一个经度/纬度范围的大致区域内，该区域大约覆盖东经10到25度和北纬55到70度。（如果你感兴趣，可以查看这里：https://en.wikipedia.org/wiki/Geography_of_Sweden#Latitude_and_longitude）

## 保存区域区域定义可以保存为YAML格式。satpy安装包中提供了一个`areas.yaml`文件（https://github.com/pytroll/satpy/blob/feature-areas-yaml/satpy/etc/areas.yaml），但如果您想保存特定的区域定义，建议在工作目录中放置一个`areas.yaml`文件，以补充或替换标准文件。或者，您可以将区域文件放置在由环境变量`PPP_CONFIG_DIR`指向的目录中。要导出您创建的区域，请使用区域的 `create_areas_def` 方法。

## 练习 4.4将您刚刚定义的区域保存到您自己的 `areas.yaml` 文件中。通过在 `scn.resample` 调用中通过名称引用它来使用保存的区域定义对您的 metop 数据进行重采样。
