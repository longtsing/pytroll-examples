## 加载一些数据
- 第一步是找到要加载的文件，并创建一个 `Scene` 实例

```python
from satpy import Scene, find_files_and_readers

files = find_files_and_readers(base_dir='/tcenas/scratch/pytroll/ex1/',
                               reader='seviri_l1b_native')

scn = Scene(filenames=files)
```

要列出可用的数据集，请运行 'available_composite_names' 方法

```python
scn.available_dataset_names()
```

然后我们选择一个复合组件进行加载并在屏幕上显示它。要获取可用复合组件的列表，请使用 `available_composite_names` 方法。

```python
print(scn.available_composite_names())
```

```python

composite = 'natural_color'
scn.load([composite])
scn.show(composite)
```

已加载的合成数据是一个来自 `xarray` 库的 `DataArray`，包含数据、像素的投影坐标以及一些元数据。

```python
scn[composite]
```

### 练习 1.1加载并显示其他合成图，例如 `雾` 和 `气团`

## 重采样数据
对 `resample` 方法进行一次简单调用即可将数据重采样到另一个区域。此处我们使用预定义区域 `euro4`，但将在另一个教程中了解如何自定义区域。其他一些预定义区域包括 `euron1`、`eurol`、`germ`。

```python
newscn = scn.resample('euro4', radius_of_influence=20000)
```

```python
newscn.show(composite)
```

## 将图像保存到磁盘`save_dataset` 方法可用于将图像保存到磁盘。支持的写入格式包括一些标准图像格式（png、jpeg）、GeoTIFF 和 NetCDF。默认情况下，将以 GeoTIFF 格式保存，文件名基于合成名称和场景的开始时间。

```python
newscn.save_dataset(composite)
```

或者，可以指定要保存到的文件名（以及通过使用正确的扩展名来指定格式）。

```python
newscn.save_dataset(composite, filename='really_cool_pic.png')
```

## 处理各个通道还可以单独加载SEVIRI通道（使用其名称或近似波长（微米））并进行处理。要获取可用通道列表，请使用 `available_dataset_names` 方法。

```python
scn.load([10.8, 12.0])
scn['t11-t12'] = scn[10.8] - scn[12.0]
scn.show('t11-t12')
```
