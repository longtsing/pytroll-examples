<h1>通过Suomi NPP上的VIIRS观测到的森林火灾烟雾和撒哈拉沙尘</h1><p>2017年10月17日，一个携带大量颗粒物和气体的气团，这些颗粒物和气体源于伊比利亚半岛大规模且破坏性的森林/草原火灾，抵达了北欧。很可能同一气团还携带了几天前西非地区爆发事件所产生的大量撒哈拉沙尘。</p><p>在对大气背景信号进行校正后，污染空气团可以通过VIIRS传感器清晰地看到。Pyspectral 可以校正背景（气候学）大气原子的瑞利散射和气溶胶的吸收。</p>    </p>

```python
from glob import glob
from satpy.scene import Scene
scn = Scene(
        filenames=glob("/data/lang/proj/safworks/adam/oktober17smoke/npp/lvl1/npp_20171017_1216_30945/*"),
        reader='viirs_sdr')
```

<h2>加载SDR数据</h2><p>现在加载生成全色锐化（使用I1波段）真彩色RGB所需的SDR数据。<br>
这里 <em>pyspectral</em> 被用来进行大气校正。</p>

```python
scn.load(['true_color'])
```

<h2>将数据重采样到地图投影区域</h2><p>在此，我们将数据重新采样到地图投影上。我们使用一个预定义的区域，称为<em>eurol</em>。为此，使用了<em>pyresample</em>和<em>pykdtree</em>。</p></p>

```python
newscn = scn.resample('europe2km')
```

<h2>显示图片</h2>那么，现在是时候使用 <em>trollimage</em> 来显示图像了。我们还使用 <em>pycoast</em> 来添加海岸线和政治边界。

```python
newscn.show('true_color', overlay={'coast_dir': '/home/a000680/data/shapes/', 'color': (255, 140, 100)})
```
