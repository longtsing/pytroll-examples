# 为 Satpy 做贡献
完成本教程后，您已经了解了Satpy所提供的功能概述。您已经看到了以下示例：
* 加载数据* 写入数据* 数据可视化* 创建复合材料* 创建自定义复合组件* 创建MPEG4视频* 将多个轨道堆叠到一张图像中* 使用 Cartopy 绘制已加载的数据* 使用GeoViews绘制已加载的数据
虽然本教程已经涵盖了很多内容，但 Satpy 还能做更多事情，甚至还有更多它无法做到的事情。这就是您作为 Satpy 项目贡献者发挥作用的地方。Satpy 是一个开源项目，由志愿者维护。我们非常感激您能提供的任何帮助。
如果您认为对 Satpy 进行某些更改或添加功能可以让您的工作更轻松，请告诉我们。如果您在使用 Satpy 时发现错误，请告诉我们。您是否有想要实现的功能或想要修复的错误，请告诉我们。
您可以在 [GitHub](https://github.com/pytroll/satpy) 上提交错误报告和功能请求。Pytroll 团队还有一个 Slack 团队，您可以在其中与我们讨论您的工作，获取使用 Satpy 的帮助，讨论 Satpy 的设计，或只是聊聊科学编程。有关如何成为贡献者的更多信息，请参见 [贡献指南](https://satpy.readthedocs.io/en/latest/dev_guide/CONTRIBUTING.html)。
